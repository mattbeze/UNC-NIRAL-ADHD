```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import joblib
from pandas import DataFrame
from sklearn.metrics import mean_absolute_error
import xlsxwriter
from sklearn.model_selection import train_test_split

sheet = pd.ExcelFile("Naive Imputation Revised.xlsx") 
```

```python
#Get sheets
sa_1y_ibis = sheet.sheet_names[3]
sa_2y_ibis = sheet.sheet_names[4]

#Grab IBIS data
sa1y_ibis_data = sheet.parse(sa_1y_ibis)
sa1y_ibis_data_features = sa1y_ibis_data.loc[:, sa1y_ibis_data.columns] 
sa1y_ibis_data_features = sa1y_ibis_data_features.drop(['ROI'], axis=1)
sa2y_ibis_data = sheet.parse(sa_2y_ibis)
sa2y_ibis_data_features = sa2y_ibis_data.loc[:, sa2y_ibis_data.columns] 
sa2y_ibis_data_features = sa2y_ibis_data_features.drop(['ROI'], axis=1)

print(sa1y_ibis_data_features.shape)
print(sa2y_ibis_data_features.shape)
print("-------")

#Grab true gilmore data
sheet = pd.ExcelFile("Scaled Gilmore Data for MAE.xlsx")
sa_1y_gilmore = sheet.sheet_names[0]
sa_2y_gilmore = sheet.sheet_names[1]

sa1y_gilmore_data = sheet.parse(sa_1y_gilmore)
sa1y_gilmore_data_features = sa1y_gilmore_data.loc[:, sa1y_gilmore_data.columns] 
sa1y_gilmore_data_features = sa1y_gilmore_data_features.drop(['ROI'], axis=1)
sa2y_gilmore_data = sheet.parse(sa_2y_gilmore)
sa2y_gilmore_data_features = sa2y_gilmore_data.loc[:, sa2y_gilmore_data.columns] 
sa2y_gilmore_data_features = sa2y_gilmore_data_features.drop(['ROI'], axis=1)

X_train_Gilmore_sa1y, X_test_Gilmore_sa1y = train_test_split(sa1y_gilmore_data_features, test_size=0.50, random_state=42)
X_train_Gilmore_sa2y, X_test_Gilmore_sa2y = train_test_split(sa2y_gilmore_data_features, test_size=0.50, random_state=42)

print(X_train_Gilmore_sa1y.shape)
print(X_train_Gilmore_sa2y.shape)
print(X_test_Gilmore_sa1y.shape)
print(X_test_Gilmore_sa2y.shape)

#Grab prediction data
sa1y_ibis_data_features.columns = X_train_Gilmore_sa1y.columns
sa2y_ibis_data_features.columns = X_train_Gilmore_sa2y.columns

sa1y_full_data = pd.concat([sa1y_ibis_data_features, X_train_Gilmore_sa1y], axis=0, ignore_index=True)
sa2y_full_data = pd.concat([sa2y_ibis_data_features, X_train_Gilmore_sa2y], axis=0, ignore_index=True)

print("-------")
print(sa1y_full_data.shape)
print(sa2y_full_data.shape)
```

```python
sa1y_mean = sa1y_full_data.mean()
sa1y_median = sa1y_full_data.median()
print(sa1y_mean.shape)
print(sa1y_median.shape)
print("-----")
sa2y_mean = sa2y_full_data.mean()
sa2y_median = sa2y_full_data.median()
print(sa2y_mean.shape)
print(sa2y_median.shape)

print(type(sa1y_mean))
```

```python
sa1y_mean = np.array(sa1y_mean)
sa1y_mean.resize(1,148)
sa1y_mean_df = pd.DataFrame(sa1y_mean)
sa1y_mean_df = sa1y_mean_df.append([sa1y_mean_df]*(X_test_Gilmore_sa1y.shape[0]-1),ignore_index=True)
print(sa1y_mean_df)

sa2y_mean = np.array(sa2y_mean)
sa2y_mean.resize(1,148)
sa2y_mean_df = pd.DataFrame(sa2y_mean)
sa2y_mean_df = sa2y_mean_df.append([sa2y_mean_df]*(X_test_Gilmore_sa2y.shape[0]-1),ignore_index=True)
print(sa2y_mean_df)

sa1y_median = np.array(sa1y_median)
sa1y_median.resize(1,148)
sa1y_median_df = pd.DataFrame(sa1y_median)
sa1y_median_df = sa1y_median_df.append([sa1y_median_df]*(X_test_Gilmore_sa1y.shape[0]-1),ignore_index=True)
print(sa1y_median_df)

sa2y_median = np.array(sa2y_median)
sa2y_median.resize(1,148)
sa2y_median_df = pd.DataFrame(sa2y_median)
sa2y_median_df = sa2y_median_df.append([sa2y_median_df]*(X_test_Gilmore_sa2y.shape[0]-1),ignore_index=True)
print(sa2y_median_df)
```

```python
#Calculate MAE
print('MAE 1y Mean {}'.format(mean_absolute_error(X_test_Gilmore_sa1y, sa1y_mean_df)))
print('MAE 2y Mean {}'.format(mean_absolute_error(X_test_Gilmore_sa2y, sa2y_mean_df)))

print('MAE 1y Median {}'.format(mean_absolute_error(X_test_Gilmore_sa1y, sa1y_median_df)))
print('MAE 2y Median {}'.format(mean_absolute_error(X_test_Gilmore_sa2y, sa2y_median_df)))
```
